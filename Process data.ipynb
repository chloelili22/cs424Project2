```python
import pandas as pd
import numpy as np
import seaborn as sns
%matplotlib inline
import matplotlib.pyplot as plt
import os
```

```python
df2 = pd.read_csv('Taxi_Trips.csv')
```

Cleaned data without NaN and removed unnecessary columns.

```python
df_clean = df2.dropna()
df_clean = df_clean.reset_index(drop=True)
taxi = df_clean.drop(columns = ['Trip End Timestamp', 'Trip Total', 'Dropoff Centroid  Location', 'Trip Seconds', 'Trip ID', 'Taxi ID', 'Dropoff Community Area', 'Dropoff Centroid  Location','Payment Type','Pickup Census Tract', 'Dropoff Census Tract', 'Tolls', 'Extras', 'Dropoff Centroid Latitude', 'Dropoff Centroid Longitude'])
taxi = taxi[(taxi['Trip Miles'] > 0)]
taxi = taxi[(taxi['Fare'] > 0)]
taxi
```

Group data by date and count thr total number of trips of each day.

```python
s = pd.to_datetime(taxi['Trip Start Timestamp'])
date_count = s.groupby(s.dt.floor('D')).size().reset_index(name='number_of_daily_trips')
date_count.rename(columns={'Trip Start Timestamp':'Date'}, inplace=True)
date_count['Date'] = date_count['Date'].dt.strftime('%Y-%m-%d')
print(date_count)
```

Added new column total number of trips into exsting dataset that we cleaned at the beginning.

```python
taxi['Trip Start Timestamp'] = pd.to_datetime(taxi['Trip Start Timestamp'])
taxi.rename(columns={'Trip Start Timestamp':'Trip Start Timestamp'}, inplace=True)
daily_avg = taxi.set_index('Trip Start Timestamp').groupby(pd.Grouper(freq='d')).mean().round(decimals = 2)
daily_trip_avg = daily_avg.reset_index()
daily_trip_avg.insert(0,'Total Trips', date_count['number_of_daily_trips'])
daily_trip_avg
os.makedirs('Desktop', exist_ok=True)  
daily_trip_avg.to_csv('Daily Taxi Trip 2021.csv')  
```
